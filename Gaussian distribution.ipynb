**Gaussian distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy.stats import multivariate_normal
```

## Standard normal distribution

The probability density for the Gaussian ditribution is
$$
    p(x)=\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$
For example,

```python
mu = 0
sigma = 0.1
s = np.random.normal(mu, sigma, 3000)
```

```python
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
               linewidth=2, color='r')
plt.show()
```

## Multivariate Gaussian distribution

The Multivariate Gaussian distribution of an $n$-dimentional $\mathbf{x}$ = $({x_1, x_2,...,x_n})$ is
$$
p(\mathbf{x}|\boldsymbol{\mu}, \mathbf\Sigma)=\frac{1}{\sqrt{(2\pi)^2|\mathbf\Sigma|}}exp\left({-\frac{1}{2}(\mathbf{x}-\boldsymbol{\mu})^T}\mathbf\Sigma^{-1}(\mathbf{x}-\boldsymbol{\mu})\right)
$$
where $\boldsymbol{\mu}$ is the $n$-dimentional mean vector and $\mathbf\Sigma$ is covariance matrix. For example,

```python
N = 100
X = np.random.normal(1, 0.5, N)
Y = np.random.normal(2, 0.5, N)
dataX = X
dataY = Y
X, Y = np.meshgrid(X, Y)
```

```python
cov = np.cov(dataX, dataY, bias=1)
print("---MEAN---")
print("X :  ", np.round(dataX.mean(), 3))
print("Y :  ", np.round(dataY.mean(), 3))
print("---COVARIANCE---")
print(cov)
```

```python
pos = np.dstack((X, Y))
mu = [dataX.mean(), dataY.mean()]
cov = np.cov(dataX, dataY, bias=1)
rv = multivariate_normal(mu, cov)
Z = rv.pdf(pos)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
fig.show()
```
