```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import seaborn as sns; sns.set()
from scipy.stats import multivariate_normal
```

## Standard normal distribution

The probability density for the **Gaussian ditribution** is
$$
    p(x)=\frac{1}{\sqrt{2\pi\sigma^2}}exp\left({-\frac{(x-\mu)^2}{2\sigma^2}}\right)
$$
For example,

```python
mu = 0
sigma = 0.1
s = np.random.normal(mu, sigma, 3000)
```

```python
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
               linewidth=2, color='r')
plt.show()
```

## Multivariate Gaussian distribution

The **Multivariate Gaussian distribution** of an $n$-dimentional $\mathbf{x}$ = $({x_1, x_2,...,x_n})$ is
$$
p(\mathbf{x}|\boldsymbol{\mu}, \mathbf\Sigma)=\frac{1}{\sqrt{(2\pi)^2|\mathbf\Sigma|}}exp\left({-\frac{1}{2}(\mathbf{x}-\boldsymbol{\mu})^T}\mathbf\Sigma^{-1}(\mathbf{x}-\boldsymbol{\mu})\right)
$$
where $\boldsymbol{\mu}$ is the **$\mathbf{n}$-dimentional mean vector** and $\mathbf\Sigma$ is  the **covariance matrix**. For example,

### Sample to Covariance

```python
N = 100
X = np.random.normal(1, 0.5, N)
Y = np.random.normal(2, 0.5, N)
dataX = X
dataY = Y
X, Y = np.meshgrid(X, Y)
```

```python
cov = np.cov(dataX, dataY, bias=1)
print("---MEAN---")
print("X :  ", np.round(dataX.mean(), 3))
print("Y :  ", np.round(dataY.mean(), 3))
print("---COVARIANCE---")
print(cov)
```

```python
pos = np.dstack((X, Y))
mu = [dataX.mean(), dataY.mean()]
cov = np.cov(dataX, dataY, bias=1)
rv = multivariate_normal(mu, cov)
Z = rv.pdf(pos)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
fig.show()
```

```python
sns.jointplot(x=dataX,
                        y=dataY, 
                        kind="kde", 
                        space=0)
```

### Covariance to Sample

```python
def get_correlated_dataset(n, dependency, mu, scale):
    latent = np.random.randn(n, 2)
    dependent = latent.dot(dependency)
    scaled = dependent * scale
    scaled_with_offset = scaled + mu
    # return x and y of the new, correlated dataset
    return scaled_with_offset[:, 0], scaled_with_offset[:, 1]
```

```python
scale = 1.0
mu = 0
cov_xx = 1.0
cov_xy = 0.5
cov_yx = 0.5
cov_yy = 1.0
cov = np.array([[cov_xx, cov_xy],
                             [cov_yx, cov_yy]])

x, y = get_correlated_dataset(400, cov, mu, scale)
plt.figure(figsize=(4, 4), dpi=80)
plt.scatter(x, y, s=6)
```

Covariance matrix is decomposited (**Cholesky Decomposition**) and resmapled  as
$$
\mathbf\Sigma=\mathbf{L}\mathbf{L}^T    \\
\mathbf{y}=\mathbf{L}\mathbf{x}
$$
$\mathbf{L}$ is the **Lower triangular matrix**, 
and it is implemented by numpy as

```python
L = np.linalg.cholesky(cov)
L
```

```python
#  Sampling by using L
# u : source vector
# y : new vector which is 

L_demo = L
# L_demo = np.array([[1.0, 0.0],
#                                      [0.0, 1.0]])
n_dim = 2
n = 100
u = np.random.normal(loc=0, scale=1, size=n_dim*n).reshape(n_dim, n)
y = np.dot(L_demo, u)
```

```python
sns.jointplot(x=u,
                        y=y, 
                        kind="kde", 
                        space=0)
```
