# Lenear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
np.set_printoptions(precision=3)
```

```python
# Sample Scatter
rate = 100
diabetes = datasets.load_diabetes()
dataX = diabetes.data[:, np.newaxis, 2][:-rate]
testX = diabetes.data[:, np.newaxis, 2][-rate:]
dataY = diabetes.target[:-rate]
testY = diabetes.target[-rate:]
dataY = dataY[:, np.newaxis]
data = np.concatenate([dataX, dataY], axis=1)
```

```python
plt.scatter(dataX, dataY,  color='b')
```

```python
cov = np.cov(data.reshape((2, -1)), rowvar=1, bias=1)
print("---MEAN---")
print("X :  ", np.round(dataX.mean(), 3))
print("Y :  ", np.round(dataY.mean(), 3))
print("---COVARIANCE---")
print(cov)
```

```python
# normalize
normX = np.array([data[i][0]-dataX.mean() for i in range(len(dataX))])
normX = normX[:, np.newaxis]
normX /= np.amax(normX)
normY = np.array([data[i][1]-dataY.mean() for i in range(len(dataY))])
normY = normY[:, np.newaxis]
normY /= np.amax(normY)
norm = np.concatenate([normX, normY], axis=1)
```

```python
plt.scatter(normX, normY,  color='b')
```

```python
cov = np.cov(norm.reshape((2, -1)), rowvar=1, bias=1)
print("---MEAN (normalized)---")
print("X :  ", np.round(normX.mean(), 3))
print("Y :  ", np.round(normY.mean(), 3))
print("---COVARIANCE (normalized)---")
print(cov)
```

## Simple Regression

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(dataX, dataY)
predY = regr.predict(testX)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(testY, predY))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(testY, predY))

# Plot outputs
plt.scatter(testX, testY,  color='black')
plt.plot(testX, predY, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

## Ridge Regression

```python
ridge = linear_model.Ridge(alpha=1.0)
```

```python
ridge.fit(dataX, dataY)
predY = ridge.predict(testX)
```

```python
# The coefficients
print('Coefficients: \n', ridge.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(testY, predY))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(testY, predY))

# Plot outputs
plt.scatter(testX, testY,  color='black')
plt.plot(testX, predY, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
